```python
#importando a biblioteca pandas atribuindo o apelido pd
#importando a biblioteca matplotlib atribuindo o apelido plt
#importando a biblioteca seaborn atribuindo o apelido sns
#importando a biblioteca numpy atribuindo o apelido np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

#Lendo o arquivo Basebanca.csv e atribuindo o dataframe á variavel DF
df = pd.read_csv("Basebanca.csv")

# Remover colunas cujos nomes começam com 'Unnamed',
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

#comando para visualizar a base de dados
display(df)
```

```python
#count = Numero de valores não nulos na coluna
#unique = Numero de clientes que compraram tênis
#top = o produto com mais relevância no database, Tenis e o Mes de fevereiro
#feq = mostra a frequencia dessas vendas
describe = df.describe()
print(describe)
```

Qual o produto mais vendido?


```python
# total de vendas por produto
#groupby agrupando a coluna produto por mês e somando
vendas_por_produto = df.groupby('Produto')['Mes'].sum()

# produto com maior soma de vendas
produto_mais_vendido = vendas_por_produto.idxmax()
quantidade_total = vendas_por_produto.max()

# Conte quantas vezes cada produto aparece (ou seja, quantas vendas de cada produto)
contagem_produtos = df['Produto'].value_counts()

# Obtenha o produto mais vendido
produto_mais_vendido = contagem_produtos.idxmax()
vendas = contagem_produtos.max()

print(f"O produto mais vendido foi '{produto_mais_vendido}' com {vendas} vendas.")
```

Qual o mês com mais vendas?


```python
# total de vendas por produto
vendas_por_mes = df.groupby('Mes')['Produto'].sum()

# produto com maior soma de vendas
produto_mais_vendido = vendas_por_mes.idxmax()
quantidade_total = vendas_por_mes.max()

# Conte quantas vezes cada produto aparece (ou seja, quantas vendas de cada produto)
contagem_produtos = df['Mes'].value_counts()

# Obtenha o produto mais vendido
mes_com_mais_vendas = contagem_produtos.idxmax()
vendas = contagem_produtos.max()

print(f"O mês com o maior número de vendas '{mes_com_mais_vendas}' com {vendas} vendas.")
```

CRIAÇÃO DAS VARIÁVEIS

```python
# Filtrando o Data Frame com os meses
# declarando variavel, que será a representação daquele mês específico no DataBase
df_janeiro = df[df['Mes'] == 'Janeiro']
df_fevereiro = df[df['Mes'] == 'Fevereiro']
df_março = df[df['Mes'] == 'Março']
df_abril = df[df['Mes'] == 'Abril']
df_maio = df[df['Mes'] == 'Maio']
df_junho = df[df['Mes'] == 'Junho']
df_julho = df[df['Mes'] == 'Julho']
df_agosto = df[df['Mes'] == 'Agosto']
df_setembro = df[df['Mes'] == 'Setembro']
df_outubro = df[df['Mes'] == 'Outubro']
df_novembro = df[df['Mes'] == 'Novembro']
df_dezembro = df[df['Mes'] == 'Dezembro']
# Contar a frequência de cada produto vendido em cada mês
produtos_mais_vendidos_fevereiro = df_fevereiro['Produto'].value_counts()
produtos_mais_vendidos_abril = df_abril['Produto'].value_counts()
produtos_mais_vendidos_agosto = df_agosto['Produto'].value_counts()
produtos_mais_vendidos_setembro = df_setembro['Produto'].value_counts()
produtos_mais_vendidos_outubro = df_outubro['Produto'].value_counts()
produtos_mais_vendidos_novembro = df_novembro['Produto'].value_counts()

print(produtos_mais_vendidos_setembro)
```

Quais os produtos mais vendidos no mês que mais vendeu?

```python
# Gerar o gráfico de barras
produtos_mais_vendidos_fevereiro.plot(kind='bar', color='red')

# Personalizar o gráfico
plt.title('Produtos Mais Vendidos em Fevereiro')
plt.xlabel('Nome dos Produtos')
plt.ylabel('Quantidade')
plt.xticks(rotation=45)
plt.tight_layout()  # Ajusta o layout para evitar cortes

# Mostrar o gráfico
plt.show()
```

```python
# Gerar o gráfico de barras
produtos_mais_vendidos_novembro.plot(kind='bar', color='red')

# Personalizar o gráfico
plt.title('Produtos Mais Vendidos em Novembro')
plt.xlabel('Nome dos Produtos')
plt.ylabel('Quantidade')
plt.xticks(rotation=45)
plt.tight_layout()  # Ajusta o layout para evitar cortes

# Mostrar o gráfico
plt.show()
```
